De .tsv a .csv

```python
import tkinter as tk
from tkinter import filedialog, messagebox
import pandas as pd
import os
```

El código implementa una interfaz gráfica (usando la biblioteca tkinter) que permite al usuario convertir archivos TSV (Tab-Separated Values) en archivos CSV (Comma-Separated Values).
El proceso se realiza mediante una ventana que presenta dos botones: uno para realizar la conversión y otro para salir de la aplicación.

```python

def convertir_tsv_a_csv():
    # Abrir cuadro de diálogo para seleccionar archivo TSV
    archivo_tsv = filedialog.askopenfilename(
        title="Selecciona un archivo TSV",
        filetypes=[("Archivos TSV", "*.tsv")]
    )
    
    if not archivo_tsv:
        messagebox.showinfo("Operación cancelada", "No se seleccionó ningún archivo.")
        return
    
    try:
        # Leer el TSV
        df = pd.read_csv(archivo_tsv, sep="\t")
        
        # Crear nombre de salida (.csv)
        archivo_csv = os.path.splitext(archivo_tsv)[0] + ".csv"
        
        # Guardar como CSV
        df.to_csv(archivo_csv, index=False)
        
        messagebox.showinfo(
            "Conversión completa",
            f"El archivo se ha convertido correctamente:\n\n{archivo_csv}"
        )
    except Exception as e:
        messagebox.showerror("Error", f"Ocurrió un problema:\n{e}")

def salir():
    """Cierra la ventana y detiene el bucle principal."""
    ventana.destroy()

# Crear ventana principal
ventana = tk.Tk()
ventana.title("Convertidor TSV → CSV")
ventana.geometry("400x240")
ventana.resizable(False, False)

# Etiqueta descriptiva
label = tk.Label(
    ventana,
    text="Selecciona un archivo TSV para convertirlo a CSV",
    font=("Segoe UI", 11),
    wraplength=360,
    justify="center"
)
label.pack(pady=20)

# Botón de acción
boton_convertir = tk.Button(
    ventana,
    text="Buscar y convertir TSV",
    font=("Segoe UI", 10, "bold"),
    bg="#0078D7",
    fg="white",
    relief="raised",
    command=convertir_tsv_a_csv
)
boton_convertir.pack(pady=10, ipadx=10, ipady=5)

# Botón salir
boton_salir = tk.Button(
    ventana,
    text="Salir",
    font=("Segoe UI", 10, "bold"),
    bg="#D9534F",
    fg="white",
    relief="raised",
    command=salir
)
boton_salir.pack(pady=10, ipadx=10, ipady=5)

# Ejecutar la interfaz
ventana.mainloop()
```
